We have taken the data shredded from the movielens kaggle dataset, and done some more ETL on it to make it easier for a ML use-case. We have joined back some of the data that was shredded as relational data back to the main movies dataset using spark's concat_ws and collect_set functions. These will enable pandas to consume the data better than relational format. 

```python
import numpy as np
import pandas as pd

ratings = pd.read_parquet('/home/jovyan/movielens_processed/ratings')
ratings.head(10)
```

```python
movie_titles_genre = pd.read_parquet("/home/jovyan/movielens_processed/movies_ml_file")
movie_titles_genre.head(10)
```

```python
ratings = ratings.merge(movie_titles_genre,on='movie_id', how='left')
ratings.head(10)
```

```python
ratings.shape
```

Average Rating

The dataset is a collection of ratings by a number of users for different movies. Let’s find out the average rating for each and every movie in the dataset.

```python
Average_ratings = pd.DataFrame(ratings.groupby('title')['rating'].mean())
Average_ratings.head(10)
```

```python
Average_ratings['Total Ratings'] = pd.DataFrame(ratings.groupby('title')['rating'].count())
Average_ratings.head(10)
```
